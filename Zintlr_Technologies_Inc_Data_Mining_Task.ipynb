**This is website/ blog data but i have good knowledge on extraxting/ parsing data from shopping cart/ housing/ranatal sites, here is the example**

**Website link:<a href="/https://www.pararius.com/apartments/amsterdam?ac=1"> click here</a>**

```python
from bs4 import BeautifulSoup
import requests
from csv import writer

url= "https://www.pararius.com/apartments/amsterdam?ac=1"
page = requests.get(url)

soup = BeautifulSoup(page.content, 'html.parser')
lists = soup.find_all('section', class_="listing-search-item")

with open('housing.csv', 'w', encoding='utf8', newline='') as f:
    thewriter = writer(f)
    header = ['Title','Sub_Title' ,'Location', 'Rooms', 'Area']
    thewriter.writerow(header)

    for list in lists:
        title = list.find('a', class_="listing-search-item__link--title").text.replace('\n', '')
        sub_title = list.find('div', class_="listing-search-item__sub-title").text.replace('\n', '')
        loc = list.find('div', class_="listing-search-item__price").text.replace('\n', '')
        rooms = list.find('li', class_="illustrated-features__item illustrated-features__item--number-of-rooms").text.replace('\n', '')
        area = list.find('li', class_="illustrated-features__item illustrated-features__item--surface-area").text.replace('\n', '')
      
        
        info = [title,sub_title, loc, rooms, area]
        thewriter.writerow(info)
```
