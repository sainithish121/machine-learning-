### **WEB SCRAPING USING "BeautifulSoup"**

**Website link:<a href="/https://www.pararius.com/apartments/amsterdam?ac=1"> click here</a>**

```python
from bs4 import BeautifulSoup
import requests
from csv import writer
from lxml.html.clean import Cleaner

url= "https://getlatka.com/saas-companies"
page = requests.get(url)
page1 = Cleaner(style=True).clean_html(page.content)
soup = BeautifulSoup(page1, 'html.parser')
lists = soup.find('table', class_="data-table_table__SwBLY")

# with open('getlatka.json' , 'w', encoding='utf8', newline='') as f:
with open('getlatka.csv' , 'w', encoding='utf8', newline='') as f:
    thewriter = writer(f)
    header = ["Name","Revenue","Funding","Valuation","Cash Flow","Founder","Team Size","Age (Yrs)","Location","Industry","Year"]
    thewriter.writerow(header)


    for list in lists.find_all('tbody'):
      rows = list.find_all('tr')
      for row in rows:
        title=row.find('a',class_='cells_link__PfQot').text
        rev=row.find('p').text
        fund=row.find_all('p')[1].text
        val=row.find_all('p')[2].text
        cashf=row.find('span').text
        fundr=row.find('a',class_="cells_name__pBrsJ").text
        teams=row.find_all('p')[3].text
        age=row.find_all('td')[8].text
        loc=row.find_all('a')[4].text
        industry =row.find_all('a')[5].text
        year=row.find_all('td')[11].text
        # print(title,rev,fund,val,cashf,fundr,teams,age,loc,industry,year)
        info = [title,rev,fund,val,cashf,fundr,teams,age,loc,industry,year]
        thewriter.writerow(info)

```
